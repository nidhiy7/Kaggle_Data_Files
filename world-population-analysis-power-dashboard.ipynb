## <div style="text-align: left; background-color:; font-family: Trebuchet MS; color:#5CB3FF; padding: 7px; line-height:1;border-radius:1px; margin-bottom: 0em; text-align: center; font-size: 25px;border-style: solid;border-color: dark green">5.1 POPULATION BY CONTINENT</div> 

```python
import plotly.express as px

fig1 = px.choropleth(df,
                     locations='Country/Territory',
                     locationmode='country names',
                     color='2022 Population',
                     color_continuous_scale='Viridis',  # Change color scale here
                     template='plotly',  # Change background template here
                     title='2022 Population')

fig1.update_layout(font=dict(size=17, family="Franklin Gothic"))

fig1.show()
```

```python
import plotly.express as px

population2022 = df.groupby(by='Continent')['2022 Population'].sum()

fig = px.pie(values=population2022.values,
             names=population2022.index,
             color_discrete_sequence=px.colors.qualitative.Set3,
             title='Population by Continent in 2022',
             template='plotly_white'  # Change background template here
             )

fig.update_traces(textinfo='label+percent+value', textfont_size=13,
                  marker=dict(line=dict(color='#100000', width=0.2)))

fig.data[0].marker.line.width = 2
fig.data[0].marker.line.color = 'gray'
fig.show()
```

## <div style="text-align: left; background-color:; font-family: Trebuchet MS; color:#5CB3FF; padding: 7px; line-height:1;border-radius:1px; margin-bottom: 0em; text-align: center; font-size: 25px;border-style: solid;border-color: dark green">5.2 GROWTH RATE</div> 

```python
import plotly.express as px

fig = px.choropleth(df,
                    locations='Country/Territory',
                    locationmode='country names',
                    color='Growth Rate',
                    color_continuous_scale='Viridis',
                    template='plotly',  # Change background template here
                    title='Growth Rate')

fig.update_layout(font=dict(size=17, family="Franklin Gothic"))

fig.show()
```

<a id="3"></a>
## <div style="text-align: left; background-color:; font-family: Trebuchet MS; color:#5CB3FF; padding: 7px; line-height:1;border-radius:1px; margin-bottom: 0em; text-align: center; font-size: 25px;border-style: solid;border-color: dark green">5.3 Top 10 Countries With Highest Population </div> 

```python
import plotly.express as px

top_most_population = df.groupby(by='Country/Territory')['2022 Population'].sum().sort_values(ascending=False).head(10)

fig = px.bar(x=top_most_population.index,
             y=top_most_population.values,
             color=top_most_population.index,
             color_discrete_sequence=px.colors.qualitative.Pastel,  # Change bar colors here
             text=top_most_population.values,
             title="Top 10 Countries with the most population",
             template='plotly_white'  # Change background template here
             )

fig.update_layout(
    xaxis_title="Countries",
    yaxis_title="Population",
    font=dict(size=15, family="Franklin Gothic"))

fig.show()
```

<a id="3"></a>
## <div style="text-align: left; background-color:; font-family: Trebuchet MS; color:#5CB3FF; padding: 7px; line-height:1;border-radius:1px; margin-bottom: 0em; text-align: center; font-size: 25px;border-style: solid;border-color: dark green">5.4 Top 10 Countries With Least Population </div> 

```python
import plotly.express as px

top_most_population = df.groupby(by='Country/Territory')['2022 Population'].sum().sort_values(ascending=False).head(10)

fig = px.bar(x=top_most_population.index,
             y=top_most_population.values,
             color=top_most_population.index,
             color_discrete_sequence=px.colors.qualitative.Pastel,  # Change bar colors here
             text=top_most_population.values,
             title="Top 10 Countries with the most population",
             template='plotly_white'  # Change background template here
             )

fig.update_layout(
    xaxis_title="Countries",
    yaxis_title="Population",
    font=dict(size=15, family="Franklin Gothic"))

fig.show()
```

## <div style="text-align: left; background-color:; font-family: Trebuchet MS; color:#5CB3FF; padding: 7px; line-height:1;border-radius:1px; margin-bottom: 0em; text-align: center; font-size: 25px;border-style: solid;border-color: dark green">5.5 Year to Year Population Change By Continent </div> 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set background color for axes and figure to white
sns.set(rc={"axes.facecolor": "white", "figure.facecolor": "white"})

data_population = df.copy()
data_population = pd.DataFrame(data_population.groupby(["Continent"])["1970 Population","1980 Population","1990 Population","2000 Population","2010 Population","2015 Population","2020 Population","2022 Population"].sum())
_, axs = plt.subplots(figsize=(20, 10))

sns.lineplot(x=data_population.T.index, y=data_population.T["Asia"], data=data_population.T, ax=axs, color="#006837", marker="o", linewidth=5, markersize=20)
sns.lineplot(x=data_population.T.index, y=data_population.T["Africa"], data=data_population.T, ax=axs, color="#66BD63", marker="o", linewidth=5, markersize=20)
sns.lineplot(x=data_population.T.index, y=data_population.T["Europe"], data=data_population.T, ax=axs, color="#F1CE08", marker="o", linewidth=5, markersize=20)
sns.lineplot(x=data_population.T.index, y=data_population.T["North America"], data=data_population.T, ax=axs, color="#15456B", marker="o", linewidth=5, markersize=20)
sns.lineplot(x=data_population.T.index, y=data_population.T["South America"], data=data_population.T, ax=axs, color="#556B15", marker="o", linewidth=5, markersize=20)
sns.lineplot(x=data_population.T.index, y=data_population.T["Oceania"], data=data_population.T, ax=axs, color="#A50026", marker="o", linewidth=5, markersize=20)

axs.set_title("Year-to-year population change of continents\n", fontsize=25)
axs.set_xlabel("\nYear", fontsize=20)
axs.set_ylabel("Population", fontsize=20)
axs.legend(["Asia", "Africa", "Europe", "North America", "South America", "Oceania"], title="Continent", edgecolor="#1c1c1c")
sns.despine(left=True, bottom=True)
plt.show()

```
